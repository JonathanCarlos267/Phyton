```python
import random

opciones = ["piedra", "papel", "tijeras"]

while True:
    while True:
        eleccion = input("\nElige piedra, papel o tijeras (o escribe salir si ya no quieres jugar): ").lower()

        if eleccion not in opciones and eleccion != "salir":
            print("OpciÃ³n no vÃ¡lida, intenta de nuevo.")
        else:
            break

    if eleccion == "salir":
        print("Â¡Gracias por jugar!")
        break

    aleatoria = random.choice(opciones)
    print("La computadora eligiÃ³:", aleatoria)

    # Determinamos el ganador
    if eleccion == aleatoria:
        print("Empate ðŸ˜º")
    elif (eleccion == "piedra" and aleatoria == "tijeras") \
        or (eleccion == "papel" and aleatoria == "piedra") \
        or (eleccion == "tijeras" and aleatoria == "papel"):
        print("Â¡Ganaste! ðŸ˜ºðŸ”¥")
    else:
        print("Perdiste ðŸ˜¿")
```

```python
#union
matutino = {20322,22183,123123,53234}
vespertino = {20322,22183,22222,11111}
total_matriculados = matutino.union(vespertino)
print(total_matriculados)
#union
total_matriculados = matutino | vespertino
print(total_matriculados)
#
```

```python
#interseccion
ca = {1234,121,666,12345}
al = {1234,111,112,444}
#con interseccion
ca_al = ca.intersection(al)
print(ca_al)
#con &
ca_al = ca & al
# interseccion usando &
ca_al = ca & al
print(ca_al)
```

```python
# Diferencia
solo_calculo = ca.difference(al)
print(solo_calculo)
#usando -
solo_calculo = ca - al
print(solo_calculo)
```
